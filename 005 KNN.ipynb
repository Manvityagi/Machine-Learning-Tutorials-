```python
def distance(a,b):
    return np.sqrt(sum((a-b)**2))

def knn(x,y,x_query,k=5):
    
    dist = []
    m = x.shape[0]
    for i in range(m):
        d = distance(x[i],x_query)
        dist.append((d,y[i]))
    #print(dist)
    dist.sort()
    dist = dist[:k]
    dist = np.array(dist)
    #print(dist)
    new_dist = np.unique(dist[:,1],return_counts=True)
    #print(new_dist)
    
    index = new_dist[1].argmax()
    pred = new_dist[0][index]
    
    return pred   
```
